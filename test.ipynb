```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
seeds = np.loadtxt('build/seed_list.csv', delimiter=',', skiprows=1)
verticies = np.loadtxt('build/vertex_list.csv', delimiter=',', skiprows=1)
edges = np.loadtxt('build/edge_list.csv', delimiter=',', skiprows=1)
```

```python
def plot_edge(edge):
    plt.plot([edge[0], edge[2]], [edge[1], edge[3]], color = 'grey', zorder=1)
```

```python
plt.figure(figsize=(5,5))
for edge in edges:
    plot_edge(edge)
plt.scatter(seeds[:, 0], seeds[:, 1], zorder = 2)
plt.scatter(verticies[:, 0], verticies[:, 1], s=10, zorder =3)
plt.show()
```
